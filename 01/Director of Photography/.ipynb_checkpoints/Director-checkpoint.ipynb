```python
# Write any import statements here

def getArtisticPhotographCount(N: int, C: str, X: int, Y: int) -> int:

  # i for index
  # t for type

  art = 0

  for (i, t) in enumerate(C):
      if(t == 'P'):

          # check left of p
          if (i >= (X * 2)):
              for x1 in range(i - X, i - Y - 1, -1):
                  if(x1 < 0):
                      break
                  if(C[x1] == 'A'):
                      for x2 in range(x1 - X, x1 - Y - 1, -1):
                          if(x2 < 0):
                              break
                          elif(C[x2] == 'B'):
                              art += 1

          # check right of p
          if (N - i >= (X * 2)):
              for x1 in range(i + X, i + Y + 1):
                  if(x1 > N - 1):
                      break
                  if(C[x1] == 'A'):
                      for x2 in range(x1 + X, x1 + Y + 1):
                          if(x2 > N - 1):
                              break
                          elif(C[x2] == 'B'):
                              art += 1


  return art
```

## 'A' approach

* my hope with this one was to decrease my solution to O(N^2) time versus O(N^3)
* this solutions passes 35 / 39 test cases

```python
art = 0

for (i, t) in enumerate(C):
    if(t == 'A' and i >= X and (N - i) >= X):
        
        Bl = 0
        Pl = 0
        Br = 0
        Pr = 0
        
        # check left of A
        for x1 in range(i - X, i - Y - 1, -1):
            if(x1 < 0):
                break
                    
            if (C[x1] == 'B'):
                Bl += 1
            elif (C[x1] == 'P'):
                Pl += 1
                    
        # check right of A
        for x1 in range(i + X, i + Y + 1):
            if(x1 > N - 1):
                break
                    
            if (C[x1] == 'B'):
                Br += 1
            elif (C[x1] == 'P'):
                Pr += 1
                
        art += ((Bl * Pr) + (Pl * Br))     
        
print(art)
```

```python
# Write any import statements here

def getArtisticPhotographCount(N: int, C: str, X: int, Y: int) -> int:

  art = 0

  for (i, t) in enumerate(C):
      if(t == 'A' and i >= X and (N - i) >= X):

          Bl = 0
          Pl = 0
          Br = 0
          Pr = 0

          # check left of A
          for x1 in range(i - X, i - Y - 1, -1):
              if(x1 < 0):
                  break

              if (C[x1] == 'B'):
                  Bl += 1
              elif (C[x1] == 'P'):
                  Pl += 1

          # check right of A
          for x1 in range(i + X, i + Y + 1):
              if(x1 > N - 1):
                  break

              if (C[x1] == 'B'):
                  Br += 1
              elif (C[x1] == 'P'):
                  Pr += 1

          art += (Bl * Pr) + (Pl * Br)     

  return art
```
